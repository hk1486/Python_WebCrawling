```python
import requests
from bs4 import BeautifulSoup
```

```python
#request = requests.get('https://www.melon.com/chart/index.htm')
#print(request) # <Response [406]> : 헤더 정보를 읽을 수 없습니다. 페이지 연결 실패

# 헤더 정보 참조
# https://zinee-world.tistory.com/512
# 헤더 정보 때문에 웹 사이트 데이터를 가져오지 못하는 경우
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
request = requests.get('https://www.melon.com/chart/index.htm',headers = header)
# print(request) # <Response [200]>

html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
titles = soup.findAll('div', {'class' : 'rank01'})
# print(titles)
artists = soup.findAll('span',{'class' , 'checkEllipsis'})
# print(artists)

"""
for i in range(len(titles)):
    title = titles[i].text.strip()
    artist = artists[i].text.strip()
    print('{0:3d}위 {1} - {2}'.format(i+1, artist, title))
"""

# 파일을 만들어서 데이터 저장
file = open('melonTop100.txt', 'w', -1, 'UTF-8')

for i in range(len(titles)):
    title = titles[i].text.strip()
    artist = artists[i].text.strip()
    data = '{0:3d}위 {1} - {2}'.format(i+1, artist, title)
    file.write(data + '\n')
print('파일 쓰기 완료!')

file.close()
```

```python
# 만들어진 파일을 읽어보자!
file = open('melonTop100.txt', 'r', -1, 'UTF-8')
lines = file.readlines()
print('멜론뮤직 TOP 100')
for line in lines:
    print(line, end='')
file.close()
```
