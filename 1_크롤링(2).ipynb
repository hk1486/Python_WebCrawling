# 벅스 뮤직 https://music.bugs.co.kr/
## 벅스 뮤직 차트 크롤링



```python
import requests
from bs4 import BeautifulSoup
```

```python
request = requests.get('https://music.bugs.co.kr/chart')
# print(request)
html = request.text
# print(html)
soup = BeautifulSoup(html,'html.parser')

titles = soup.findAll('p', {'class' : 'title'})
# print(titles)
artists = soup.findAll('p', {'class' : 'artist'})
# print(artists)

for i in range(len(titles)):
    title = titles[i].text.strip() # strip():문자열 앞뒤의 불필요한 빈칸을 제거
    # print(title)
    artist = artists[i].text.strip().split('\n')[0] # split()[0] : 괄호안 내용을 제거하고 앞 데이터만 출력
    # print(artist)
    data = '{0:3d}위 {1} - {2}'.format(i+1, artist, title)
    print(data)
```
