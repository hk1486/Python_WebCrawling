### 머신러닝 프로세스
- 기획 -> 데이터 수집 -> 데이터 탐색 -> 데이터 전처리 -> 모델링 -> 모델평가 -> 결과보고서

#### 데이터 수집 -> 크롤링(스크레이핑)
##### 웹 크롤러(Web Crawler)
- 자동화된 방법으로 웹에서 다양한 정보를 수집하는 소프트웨어를 말합니다.
- HTML(CSS, JAVASCRIPT)문서를 가져오려면 웹 사이트의 URL로 접속하여 데이터를 가져옵니다. (requests 모듈)
- HTML문서를 파싱(분석)하려면 BeautifulSoup 모듈을 설치해야 합니다.

```python
# pip install requests
# pip install beautifulsoup4

import requests
from bs4 import BeautifulSoup

# https://basicenglishspeaking.com/daily-english-conversation-topics/
```

```python
def getSubject():
    # print('getSubject() 함수 호출')
    subjects = [] # 파싱된 데이터를 저장할 빈 리스트를 만듭니다.
    request = requests.get('https://basicenglishspeaking.com/daily-english-conversation-topics/') 
    print(request) # 리스펀스 200 => 정상적인 연결.
    html = request.text # 요청한 페이지 정보중에서 문자열만 저장.
    # print(html)
    # BeautifulSoup(파싱할 HTML소스, 파싱에 사용될 파서)
    # html 문서를 파싱 : html.parser / xml문서를 파싱 : xml.parser
    soup = BeautifulSoup(html,'html.parser')
    # print(type(soup))
    divs = soup.findAll('div',{"class":'thrv-columns'})
    # print(divs)
    
    for div in divs:
        links = div.findAll('a')
        #print(links)
        for link in links:
            # print(link.text)
            subjects.append(link.text)
    # print(subjects)
    return subjects


if __name__ == "__main__": #파이썬 프로그램 시작점을 만듭니다.
    # print('파이썬 프로그램의 시작')
    result = getSubject()
    # print(result)
    print('총', len(result),'개의 대화를 찾았습니다')

    for i in range(len(result)):
        print('{0:2d}, {1:s}'.format(i+1, result[i]))
```
